```python
import numpy as np
import matplotlib.pyplot as plt

# Define ReLU, sigmoid, and their derivatives
def relu(z):
    return np.maximum(0, z)

def relu_prime(z):
    return np.where(z >= 0, 1, 0)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_prime(z):
    sigmoid_z = sigmoid(z)
    return sigmoid_z * (1 - sigmoid_z)
```

```python
# Generate input values
x_values = np.linspace(-5, 5, 100)

# Compute function values
relu_values = relu(x_values)
relu_prime_values = relu_prime(x_values)

sigmoid_values = sigmoid(x_values)
sigmoid_prime_values = sigmoid_prime(x_values)
```

```python
# Plot ReLU and its derivative
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x_values, relu_values, label='ReLU')
plt.title('ReLU Activation Function')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x_values, relu_prime_values, label='ReLU Derivative', color='orange')
plt.title('ReLU Derivative')
plt.legend()
```

```python
# Plot sigmoid and its derivative
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x_values, sigmoid_values, label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x_values, sigmoid_prime_values, label='Sigmoid Derivative', color='orange')
plt.title('Sigmoid Derivative')
plt.legend()

plt.tight_layout()
plt.show()
```
